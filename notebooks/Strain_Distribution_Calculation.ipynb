# Strain Distribution Calculation

This notebook demonstrates the packaged strain workflow. The element strain and point-smoothed strain calculations are implemented in `src/pointcloud_etfe_postprocessing/strain.py`.

```python
from pathlib import Path
import sys

import pandas as pd
from IPython.display import Image, display

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

src_path = PROJECT_ROOT / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

DATA_DIR = PROJECT_ROOT / "data" / "raw"
OUT_DIR = PROJECT_ROOT / "outputs" / "notebooks"
print(f"Project root: {PROJECT_ROOT}")
print(f"Data directory: {DATA_DIR}")
```

## Run the workflow

```python
from pointcloud_etfe_postprocessing.config import TriangulationConfig
from pointcloud_etfe_postprocessing.workflows import run_strain_workflow

result = run_strain_workflow(
    DATA_DIR / "zxt_300Pa.xlsx",
    DATA_DIR / "zxt_9000Pa.xlsx",
    OUT_DIR / "strain",
    triangulation=TriangulationConfig(method="structured", boundary_scale=1.6),
    plot=True,
)
result.paths
```

## Inspect point-level strain

```python
point_strain_csv = OUT_DIR / "strain" / "zxt_9000Pa_point_strain.csv"
point_strain = pd.read_csv(point_strain_csv)
print(f"Rows: {len(point_strain)}")
point_strain[["epsilon1", "epsilon2"]].agg(["min", "mean", "max"]).round(5)
```

```python
point_strain.head()
```

## Inspect element-level strain

```python
element_strain_csv = OUT_DIR / "strain" / "zxt_9000Pa_element_strain.csv"
element_strain = pd.read_csv(element_strain_csv)
print(f"Rows: {len(element_strain)}")
element_strain[["E11", "E22", "E12", "epsilon1", "epsilon2", "tau"]].agg(["min", "mean", "max"]).round(5)
```
